```python
df_jan['tpep_dropoff_datetime'] = pd.to_datetime(df_jan['tpep_dropoff_datetime'])
df_jan['tpep_pickup_datetime'] = pd.to_datetime(df_jan['tpep_pickup_datetime'])
```

```python
df_jan['duration'] = (df_jan['tpep_dropoff_datetime'] - df_jan['tpep_pickup_datetime'])/pd.Timedelta(minutes=1)
```

```python
df_jan = df_jan[(df_jan['duration']<=60) & (df_jan['duration']>=1)]
```

```python
cat_features = ['PULocationID','DOLocationID']
df_jan[cat_features] = df_jan[cat_features].astype('str')
df_jan.dtypes
```

```python
from sklearn.feature_extraction import DictVectorizer
traindict = df_jan[cat_features].to_dict(orient='records') 
```

```python
dv = DictVectorizer()
xtrain = dv.fit_transform(traindict)
```

```python
xtrain
```

```python
target = df_jan['duration'].values
ytrain = target
```

```python
ytrain
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr = LinearRegression()
lr.fit(xtrain, ytrain)
```

```python
ypred_train = lr.predict(xtrain)
rmse = mean_squared_error(y_true=ytrain,y_pred=ypred_train,squared=False) 
print(f' Train RMSE is: {rmse}')
```

```python
df_feb = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-02.parquet')
```

```python
df_feb['tpep_dropoff_datetime'] = pd.to_datetime(df_feb['tpep_dropoff_datetime'])
df_feb['tpep_pickup_datetime'] = pd.to_datetime(df_feb['tpep_pickup_datetime'])
df_feb['duration'] = (df_feb['tpep_dropoff_datetime'] - df_feb['tpep_pickup_datetime']) / pd.Timedelta(minutes=1)
df_feb = df_feb[(df_feb['duration'] <= 60) & (df_feb['duration'] >= 1)]
df_feb[cat_features] = df_feb[cat_features].astype('str')
testdict = df_feb[cat_features ].to_dict(orient='records') 
xtest = dv.transform(testdict)
target = df_feb['duration'].values
ytrain = target
```

```python
ypred_test = lr.predict(xtest)
rmse = mean_squared_error(y_true=ytrain,y_pred=ypred_test,squared=False) 
print(f' Test RMSE is: {rmse}')
```
